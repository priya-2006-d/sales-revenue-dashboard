**Task 2-Load Data**

```python
import pandas as pd

df = pd.read_csv('train.csv')
df.head()
```

**Task 3-Check Column**

```python
print(df.columns)
```

**Task 4-Total Sales KPI**

```python
total_sales = df['Sales'].sum()
print("Total Sales:", total_sales)
```

**Task 5-Sales by State**

```python
import matplotlib.pyplot as plt

state_sales = df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
state_sales.plot(kind='bar')
plt.title('Top 10 States by Sales')
plt.show()
```

**Task 6-Sales by Segment**

```python
segment_sales = df.groupby('Segment')['Sales'].sum()

plt.figure(figsize=(6,6))
segment_sales.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.title('Sales by Segment')
plt.show()
```

**Task 7-Top 10 Products**

```python
product_sales = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
product_sales.plot(kind='bar')
plt.title('Top 10 Products')
plt.show()
```

**Task 8-Sales Trend**

```python
df['Order ID.1'] = pd.to_datetime(df['Order ID.1'], format='%d-%m-%Y')

sales_trend = df.groupby('Order ID.1')['Sales'].sum()

plt.figure(figsize=(12,5))
sales_trend.plot()
plt.title('Sales Trend Over Time')
plt.show()
```
